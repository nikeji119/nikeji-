# simple linear regression section

```python
# using 'soft_living' to predict price
x_sqrt = trans_df.iloc[:, 2] # picking all the rows and index 2
y_sqrt = trans_df.iloc[:, -1] # picking all the rrows and column on index -1
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_sqrt, y_sqrt, test_size = .2)
```

```python
x_train = np.array(x_train).reshape(-1,1)
x_test = np.array(x_test).reshape(-1,1)
```

```python
# building  our model - simple linear regression model btw saft

taiwo = LinearRegression()
taiwo.fit(x_train, y_train)

target = taiwo.predict(x_test)
```

```python
print('MSE:', mean_squared_error(y_test, target))

print('R squared:', metrics.r2_score(y_test, target))
```

```python
# using 'yr_renovated' to predict price
x_yr_ren = trans_df.iloc[:, 11] # picking all the rows and index 2
y_yr_ren = trans_df.iloc[:, -1] # picking all the rrows and column on index -1
x_train_yr_ren, x_test_yr_ren, y_train_yr_ren, y_test_yr_ren = train_test_split(x_yr_ren, y_yr_ren, test_size = .2)
x_train_yr_ren = np.array(x_train_yr_ren).reshape(-1,1)
x_test_yr_ren = np.array(x_test_yr_ren).reshape(-1,1)



yr_ren= LinearRegression()
yr_ren.fit(x_train_yr_ren, y_train_yr_ren)

target = yr_ren.predict(x_test_yr_ren)

print('MSE:', mean_squared_error(y_test_yr_ren, target))

print('R squared:', metrics.r2_score(y_test_yr_ren, target))
```

```python
# using 'yr_built' to predict price
x_yr_built = trans_df.iloc[:, 10] # picking all the rows and index 2
y_yr_built = trans_df.iloc[:, -1] # picking all the rrows and column on index -1
x_train_yr_built, x_test_yr_built, y_train_yr_built, y_test_yr_built = train_test_split(x_yr_built, y_yr_built, test_size = .2)
x_train_yr_built = np.array(x_train_yr_built).reshape(-1,1)
x_test_yr_built = np.array(x_test_yr_built).reshape(-1,1)



year_built = LinearRegression()
year_built.fit(x_train_yr_built, y_train_yr_built)

target = year_built.predict(x_test_yr_built)

print('MSE:', mean_squared_error(y_test_yr_built, target))

print('R squared:', metrics.r2_score(y_test_yr_built, target))
```

```python
# using 'lat' to predict price
x_lat = trans_df.iloc[:, 11] # picking all the rows and index 2
y_lat = trans_df.iloc[:, -1] # picking all the rrows and column on index -1
x_train_lat, x_test_lat, y_train_lat, y_test_lat = train_test_split(x_lat, y_lat, test_size = .2)
x_train_lat = np.array(x_train_lat).reshape(-1,1)
x_test_lat = np.array(x_test_lat).reshape(-1,1)



yr_built= LinearRegression()
yr_built.fit(x_train_lat, y_train_lat)

target = yr_built.predict(x_test_lat)

print('MSE:', mean_squared_error(y_test_lat, target))

print('R squared:', metrics.r2_score(y_test_lat, target))
```
