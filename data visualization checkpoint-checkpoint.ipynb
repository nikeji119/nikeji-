```python
old_title = []
new_title = []

for name in titanic_2['Name']:
    for t,u in zip(T_dic.keys(),T_dic.values()):
        t = t +'.'
        if t in name:
            old_title.append(t)
            new_title.append(u)
```

```python
len(old_title)==len(titanic_2)
```

```python
titanic_2['old_title'] = old_title
titanic_2['new_title'] = new_title
```
